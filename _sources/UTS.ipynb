```python
import pandas as pd
import re
import string
import numpy as np
import nltk
import swifter
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import matplotlib.pyplot as plt
```

```python
def case_folding(comment):
    comment = comment.lower()
    return comment
def linkNormalize(text):
    text = re.sub(r"\s—\s", "", text)
    text = re.sub(r"http\S+", "", text)
    return text
symbols = ",!\"#$%&()*+-.…/:;<=>?@[\]^_`{|}~\n"
def cleaning_simbol(comment):
    for i in symbols:
        comment = np.char.replace(comment, i, ' ')
    return comment
def tokenize(comment):
    return comment.split()

nltk.download('stopwords')
list_stopwords = stopwords.words('indonesian')
def stopword_removal(comment):
    return [word for word in comment if word not in list_stopwords]

factory = StemmerFactory()
stemmer = factory.create_stemmer()
def stemmingNormalize(comment):
#     print(comment)
    kata = []
    for term in comment:
        term = stemmer.stem(term)
        kata.append(term)
    sentence = ""
    for i in kata:
      sentence += i
      sentence += " "
    return sentence
```

### 1. Case - Folding Data
> Proses case folding bertujuan untuk membuat seluruh kata menjadi huruf kecil semua.



```python
X = X.apply(case_folding)
k=0
for i in X:
    #print(k,i)
    k += 1
```

### 2. Cleaning - Text Data


> 1. linkNormalize, bertujuan untuk menghilangkan alamat link pada data
> 2. cleaning_simbol, bertujuan untuk menghilangkan simbol yang tidak perlu pada data





```python
X = X.apply(linkNormalize)
k=0
for i in X:
    #print(k,i)
    k += 1
```

```python
X = X.apply(cleaning_simbol)
k=0
for i in X:
    text = i.split()
    #print(k,i)
    k += 1
```

### 3. Tokenize
> Proses tokenizing bertujuan untuk memecah kalimat menjadi kata.



```python
X = X.apply(tokenize)
X
```

### 4. Stopword Removal
> Proses stopword removal bertujuan untuk menghilangkan kata yang tidak perlu.



```python
X = X.apply(stopword_removal)
k=0
for i in X:
    #print(k,i)
    k += 1
```

### 5. Stemming
> Proses stemming bertujuan untuk membuat kata yang berimbuhan menjadi kata dasar.



```python
X = X.swifter.apply(stemmingNormalize)
X
```

## **3. Proses Representasi Data Menggunakan Algoritma TF-IDF**


> Term Frequency — Inverse Document Frequency atau TF — IDF adalah suatu metode algoritma yang berguna untuk menghitung bobot setiap kata yang umum digunakan. Metode ini akan menghitung nilai Term Frequency (TF) dan Inverse Document Frequency (IDF) pada setiap token (kata) di setiap dokumen. Secara sederhana, metode TF-IDF digunakan untuk mengetahui berapa sering suatu kata muncul di dalam dokumen.



```python
from sklearn.feature_extraction.text import TfidfVectorizer
# )
tfidf_vectorizer = TfidfVectorizer()
tfidf_separate = tfidf_vectorizer.fit_transform(X)

df_tfidf = pd.DataFrame(
    tfidf_separate.toarray(), columns=tfidf_vectorizer.get_feature_names_out(), index=df.index
)
X = df_tfidf.values
df_tfidf
```

## **4. Proses Splitting Data**

> Dalam proses splitting data ini dibagi menjadi 2 bagian, yakni data training dan data testing. Data training menggunakan sebanyak 80% dari keseluruhan data sedangkan data testing menggunakan sebanyak 20% dari keseluruhan data



```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( 
    X, Y, test_size = 0.2, random_state = 100)

y_train
```

## **5. Proses Reduksi Data Menggunakan Metode PCA dan Proses Klasifikasi Data menggunakan metode KNN (K-Nearest Neighbor)**

> Principal Component Analysis (PCA) adalah salah satu metode reduksi dimensi pada machine learning. PCA akan memilih œvariabel-variabel yang mampu menjelaskan sebagian besar variabilitas data. PCA mengurangi dimensi dengan membentuk variabel-variabel baru

> KNN merupakan algoritma klasifikasi yang paling sederhana dalam mengklasifikasikan sebuah gambar kedalam sebuah label. Metode ini mudah dipahami dibandingkan metode lain karena mengklasifikasikan berdasarkan jarak terdekat dengan objek lain (tetangga).





### Tahap Uji Coba

1. Uji coba pengecekan hasil akurasi data menggunakan metode klasifikasi KNN (K-Nearest Neighbor) terhadap data **sebelum** dilakukannya proses reduksi data menggunakan metode PCA.

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(X_test)
print (" KNN Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
```


> **Sebelum** dilakukannya proses reduksi data menggunakan metode PCA, pengecekan hasil akurasi data menggunakan metode klasifikasi KNN dengan jumlah fitur data sebanyak **6570** menghasilkan akurasi sebesar **83.85**.



2. Uji coba pengecekan hasil akurasi data menggunakan metode klasifikasi KNN (K-Nearest Neighbor) terhadap data **setelah** dilakukannya proses reduksi data menggunakan metode PCA.
> 



```python
from sklearn.decomposition import PCA

pca = PCA(n_components =10)

X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)

pca.explained_variance_ratio_
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(X_test)
print (" KNN Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
```


> **Setelah** dilakukannya proses reduksi data menggunakan metode PCA dengan jumlah n_components sebanyak 10, pengecekan hasil akurasi data menggunakan metode klasifikasi KNN dengan jumlah fitur data sebanyak **10** menghasilkan akurasi sebesar **82.60**.



## **Kesimpulan**

> Dari tahap uji coba yang dilakukan pada proses reduksi data menggunakan metode PCA dan proses klasifikasi data menggunakan metode KNN (K-Nearest Neighbor), dapat disimpulkan bahwa dengan diterapkannya metode PCA dapat mempercepat proses komputasi klasifikasi karena data yang sebelumnya memiliki fitur 6570 menjadi 10 fitur. Meskipun hasil akurasi turun sebanyak **83.85 - 82.60 = 1.25**, namun dapat mempercepat proses komputasi klasifikasi.



## **6. Proses Reduksi Data Menggunakan Metode PCA dan Proses Klasifikasi Data menggunakan metode K-Means (Jumlah Cluster 2)**

> K-means clustering dapat didefinisikan sebagai algoritma iteratif yang membagi kumpulan data (dataset) yang tidak berlabel menjadi k kluster yang berbeda sedemikian rupa sehingga setiap kumpulan data hanya dimiliki oleh satu kelompok yang memiliki properti serupa.




```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
```

```python
center = kmeans.cluster_centers_
center
```

```python
y_pred = kmeans.labels_


data = kmeans.labels_
res = []
for i in data:
    if i == 0:
        res.append("RPL")
    else:
        res.append("KOMPUTASI")
print(res)
```

```python
y_true = df['Kategori'].values.tolist()
print(df['Kategori'].values.tolist())
```

```python
n = len(y_true)
for i in range(n):
    print(res[i],"\t : ",y_true[i])
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ",
    accuracy_score(y_true, res)*100)
```
