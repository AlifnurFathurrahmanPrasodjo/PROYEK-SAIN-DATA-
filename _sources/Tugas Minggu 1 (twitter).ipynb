```python
import pandas as pd
import snscrape.modules.twitter as sntwitter

# Atur parameter pencarian
keyword = "Ridwan kamil"
max_tweets = 500

# Buat list kosong untuk menyimpan tweet
tweets_list = []

# Lakukan pencarian tweet
for i, tweet in enumerate(sntwitter.TwitterSearchScraper(keyword + " lang:id since:2021-01-01 until:2023-02-28").get_items()):
    if i > max_tweets:
        break
    tweets_list.append([tweet.username, tweet.content])

# Konversi list ke pandas dataframe
df = pd.DataFrame(tweets_list, columns=["username", "tweet"])

# Simpan hasil ke dalam file CSV
df.to_csv("data-tweet.csv", index=False)

print("Data berhasil disimpan ke dalam file data-tweet.csv")
```
